```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Environmental indicator columns
columns_check = [
    'mean_ghgs', 'mean_land', 'mean_watscar', 'mean_eut',
    'mean_ghgs_ch4', 'mean_ghgs_n2o', 'mean_bio',
    'mean_watuse', 'mean_acid'
]

# Create a multi-plot layout for boxplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(columns_check):
    sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot - {col}')

plt.tight_layout()
plt.suptitle("Boxplots of Environmental Indicators", fontsize=16, y=1.02)
plt.show()

# KDE + Histogram plots (displaying only the first few indicators for brevity)
for col in columns_check[:4]:  # You can change this to all, or select key columns
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='coral')
    plt.title(f'Distribution of {col} (KDE + Histogram)')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

# Scatterplot: Check for potential outliers or edge cases in selected variable pairs
plt.figure(figsize=(6, 5))
sns.scatterplot(x=df['mean_ghgs'], y=df['mean_land'], hue=df['diet_group'])
plt.title('Scatterplot: GHG Emissions vs Land Use')
plt.xlabel('mean_ghgs')
plt.ylabel('mean_land')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 5))
sns.scatterplot(x=df['mean_ghgs'], y=df['mean_watscar'], hue=df['diet_group'])
plt.title('Scatterplot: GHG Emissions vs Water Scarcity')
plt.xlabel('mean_ghgs')
plt.ylabel('mean_watscar')
plt.tight_layout()
plt.show()
```

```python
# handle outliers using IQR method
columns_with_outliers = ['mean_watscar', 'mean_bio', 'mean_watuse']

for col in columns_with_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out outliers based on IQR bounds
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    # Print the number of remaining rows after filtering
    print(f"{col}: Number of rows retained = {df.shape[0]}")
```

```python
# Visualize the data after outlier removal
sns.boxplot(data=df[columns_with_outliers])
plt.xticks(rotation=45)
plt.title("After Outlier Removal")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px

# Get numerical columns
numerical_cols = df.select_dtypes(include='number').columns.tolist()

# Aggregate: compute the mean only for numerical columns, grouped by diet group
df_avg = df.groupby('diet_group')[numerical_cols].mean().reset_index()

# Normalize values (excluding 'diet_group') using MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_avg[numerical_cols])
scaled_df = pd.DataFrame(scaled, columns=numerical_cols)
scaled_df['diet_group'] = df_avg['diet_group']
scaled_df = scaled_df[['diet_group'] + numerical_cols]  # Ensure correct column order

# Select and order key environmental indicators
ordered_indicators = [
    'mean_ghgs', 'mean_ghgs_ch4', 'mean_ghgs_n2o',
    'mean_land', 'mean_watscar', 'mean_watuse',
    'mean_eut', 'mean_acid', 'mean_bio'
]

# Reshape the DataFrame to long format for plotting
df_long = scaled_df.melt(id_vars='diet_group', var_name='Indicator', value_name='Value')
df_long = df_long[df_long['Indicator'].isin(ordered_indicators)]

# Friendly display names for indicators (for axis labels)
indicator_labels = {
    'mean_ghgs': 'GHG Emissions (kg)',
    'mean_ghgs_ch4': 'CH₄ Emissions (kg)',
    'mean_ghgs_n2o': 'N₂O Emissions (kg)',
    'mean_land': 'Land Use (m²)',
    'mean_watscar': 'Water Scarcity',
    'mean_watuse': 'Water Use (m³)',
    'mean_eut': 'Eutrophication (gPO₄eq)',
    'mean_acid': 'Acidification',
    'mean_bio': 'Biodiversity Impact (extinction/day)'
}

df_long['Indicator'] = pd.Categorical(df_long['Indicator'], categories=ordered_indicators, ordered=True)
df_long = df_long.sort_values(['diet_group', 'Indicator'])

# Define custom color mapping for diet groups
color_map = {
    'Vegan': '#145A32',        
    'Veggie': '#2ECC71',        
    'Fish': '#3498DB',          
    'Meat50': '#F39C12',        
    'Meat': '#E74C3C',          
    'Meat100': '#8B0000'        
}

# Plot radar chart using friendly labels and custom colors
fig = px.line_polar(
    df_long,
    r='Value',
    theta='Indicator',
    color='diet_group',
    color_discrete_map=color_map,
    line_close=True,
    template='plotly_white',
    markers=True,
    title='Environmental Impact of Different Diet Types',
    labels={"Indicator": "Environmental Indicator"}
)

# Update axis labels using descriptive indicator names
fig.update_traces(
    theta=[indicator_labels[ind] for ind in df_long['Indicator']]
)

fig.show()
```

```python
# Update layout styling: font, title, axis ticks, legend position
fig.update_layout(
    font=dict(
        family="Arial",         
        size=28,               
        color="black"           
    ),
    title_font=dict(
        size=36                 
    ),
    polar=dict(
        radialaxis=dict(
            tickfont_size=24,  
            gridcolor="gray",   
            linecolor="gray"    
        ),
        angularaxis=dict(
            tickfont_size=28    
        )
    ),
    legend=dict(
        font=dict(size=26),     
        x=1.05,                 
        y=1                    
    )
)

# Export the radar chart as a high-resolution PNG image
fig.write_image("radar_chart.png", width=1900, height=1000, scale=3)
```
